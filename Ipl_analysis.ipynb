### What was the number of matches played in each year?

Number of matches played in various season:

```python
Match_data['Season']=pd.DatetimeIndex(Match_data['date']).year
#Match_data.head()
```

```python
match_per_season=Match_data.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
Match_data.nlargest(5,'Season')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Season',y='matches',data=match_per_season)
plt.show()
```

```python
seas_dat=Match_data[['id','Season']].merge(ball_by_ball,left_on='id',right_on='id',how='left').drop('id',axis=1)
seas_dat.head()
```

```python
season=seas_dat.groupby(['Season'])['total_runs'].sum().reset_index()
season
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(x='Season',y='total_runs',data=season)
plt.xlabel("Season")
plt.ylabel("Total runs")
plt.title("Total runs across the seasons",fontsize=24)

plt.show()
```

```python
season.nlargest(5,'total_runs')
```

```python
runs=pd.concat([match_per_season,season.iloc[:,1]],axis=1)
runs
```

```python
#sns.lineplot(x='Season',y='matches',data=runs)

runs['Runs per Match']=runs['total_runs']/runs['matches']
runs.set_index('Season',inplace=True)
runs.style.background_gradient(cmap='PuBu',subset=['Runs per Match'])
```

```python
plt.plot(runs.index,runs['Runs per Match'],color='green')

plt.title("Runs per Match across Seasons ",fontsize=16)
plt.xlabel("Seasons")
plt.ylabel("Runs per match")
plt.grid()
plt.show()
```

```python
temp=Match_data.toss_decision.value_counts()
temp
```

```python
labels=temp.index
val=(temp/temp.sum())*100
plt.figure(figsize=(8,6))
plt.pie(val,labels=labels,radius=0.7,autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=Match_data,x='Season',hue='toss_decision',multiple='stack')
plt.show()
```
